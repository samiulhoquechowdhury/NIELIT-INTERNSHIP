```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_sal = pd.read_csv('Salary_Data.csv')
df_sal.head()
```

```python
plt.title('Distribution of the dependent variable')
sns.histplot(df_sal['Salary'])
plt.show()
```

```python
plt.title('Years of Experience vs Salary')
sns.scatterplot(x='YearsExperience',y='Salary',data=df_sal,color='blue')
```

```python
#Select our independent variable
#Select all rows and 1st column from the data
x=df_sal.iloc[:,:1]
```

```python
#Select our dependent variable
#Select all rows and 2nd column from the data
y=df_sal.iloc[:,1:]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
regressor = LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
y_prediction=regressor.predict(xtest)
y_prediction
```

```python
y_predictionTrain=regressor.predict(xtrain)
y_predictionTrain
```

```python
plt.scatter(xtest,y_prediction,color = 'blue')
plt.plot(xtrain,y_predictionTrain, color='red')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
#Scatter plot for xtrain and ytrain
#Plot a line for xtrain and its prediction
#Give title as 'Salary vs Experience (Training Set)'
```

```python
mse=mean_squared_error(ytest,y_prediction) #Pass the actual value and predicted value for the test data
r2=r2_score(ytest,y_prediction)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
